LangGraph organizes workflows as graphs where nodes are functions and edges define execution flow. All nodes share a common state that gets passed between them. This notebook shows how to define state, create nodes, and connect them into an executable graph.


```python
from IPython.display import Image, display
import operator
from typing import Annotated, List, Literal, TypedDict
from langgraph.graph import END, START, StateGraph
from langgraph.types import Command, interrupt
```

<a id='state_definition'></a>

```python
class State(TypedDict):
    nlist: List[str]
```

A TypedDict in Python is a way to describe what keys and values a dictionary should have. Normally, a dictionary can hold anything, with any keys and any types of values. TypedDict lets you define a ‚Äúblueprint‚Äù so that tools like type checkers and IDEs know exactly what keys exist and what type of values they should contain.
In your example:

''class State(TypedDict):
    nlist: List[str]''

This means you are defining a dictionary type called State. It must have a key named "nlist", and the value for that key must be a list of strings. For example:

Why use TypedDict? It makes your code clearer, safer, and easier to work with. It tells you and your tools exactly what kind of dictionary you expect, so you catch mistakes early and get autocomplete support.
The simple idea is: TypedDict says ‚Äúthis dictionary must look like this, with these keys and these types.‚Äù It‚Äôs a way to give structure to something that is usually very flexible.


<a id='node_function'></a>


```python
def node_a(state: State) -> State:
    print(f"node a is receiving {state['nlist']}")
    note = "Hello World from Node a"
    return(State(nlist = [note]))
```

```python
builder = StateGraph(State)
builder.add_node("a", node_a)
builder.add_edge(START, "a")
builder.add_edge("a", END)
graph = builder.compile()
```

```python
display(Image(graph.get_graph().draw_mermaid_png()))
# print(graph.get_graph().draw_mermaid())
```

<a id='graph_invoke'></a>

```python
initial_state = State(
    nlist = ["Hello Node a, how are you?"]
)
graph.invoke(initial_state)
```

## Takeaways

Setup:

- State: All nodes share the same [state](#state_definition) which can be a Python TypedDict, dataclass, or a Pydantic BaseModel
- Nodes: Defined as simple [Python functions](#node_function) that receive state as input and return updated state

Execution (invoke):

- Runtime: When you call [invoke](#graph_invoke), the graph initializes the input state from your invoke statement and determines which nodes to run
- State Flow: Each node receives the current state as input, executes its logic, and returns an updated state
- Graph Return: After all nodes complete execution, the graph returns the final state value
